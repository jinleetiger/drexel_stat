```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample baseball data
data = {
    'Name': ['Player' + str(i) for i in range(200)],
    'Pos': ['C']*20 + ['1B']*20 + ['2B']*20 + ['3B']*20 + ['SS']*20 + 
           ['LF']*20 + ['CF']*20 + ['RF']*20 + ['DH']*20 + ['P']*20,
    'OPS': np.concatenate([
        np.random.normal(0.720, 0.050, 20),  # C
        np.random.normal(0.800, 0.050, 20),  # 1B
        np.random.normal(0.750, 0.050, 20),  # 2B
        np.random.normal(0.770, 0.050, 20),  # 3B
        np.random.normal(0.740, 0.050, 20),  # SS
        np.random.normal(0.780, 0.050, 20),  # LF
        np.random.normal(0.760, 0.050, 20),  # CF
        np.random.normal(0.790, 0.050, 20),  # RF
        np.random.normal(0.810, 0.050, 20),  # DH
        np.random.normal(0.300, 0.050, 20)   # P
    ])
}

# Create DataFrame
df = pd.DataFrame(data)

# Create the plot
plt.figure(figsize=(15, 8))

# Define position order (excluding pitchers)
pos_order = ['C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'DH']

# Filter out pitchers and create boxplot
plot_data = df[df['Pos'] != 'P']
sns.boxplot(data=plot_data, x='Pos', y='OPS', order=pos_order, palette='Set3')

# Customize the plot
plt.title('OPS Distribution by Position (Sample Data)', fontsize=14, pad=20)
plt.xlabel('Position', fontsize=12)
plt.ylabel('OPS (On-base Plus Slugging)', fontsize=12)

# Add position labels with descriptions
position_labels = {
    'C': 'Catcher',
    '1B': 'First Base',
    '2B': 'Second Base',
    '3B': 'Third Base',
    'SS': 'Shortstop',
    'LF': 'Left Field',
    'CF': 'Center Field',
    'RF': 'Right Field',
    'DH': 'Designated\nHitter'
}

# Set x-tick labels with full position names
plt.xticks(range(len(pos_order)), 
           [f"{pos}\n({position_labels[pos]})" for pos in pos_order],
           rotation=0)

# Add grid for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# Adjust layout
plt.tight_layout()

plt.show()

# Print summary statistics
print("\nSummary statistics by position:")
print(plot_data.groupby('Pos')['OPS'].describe())
```

## Stolen Base Analysis

In baseball, stolen bases are an important aspect of the running game. We'll analyze:
1. Stolen base totals by position
2. Success rates in stealing attempts
3. Relationship between speed and stolen bases
4. Stolen base efficiency (SB per plate appearance)

```python
# Create enhanced baseball data including stolen base stats
data = {
    'Name': ['Player' + str(i) for i in range(200)],
    'Pos': ['C']*20 + ['1B']*20 + ['2B']*20 + ['3B']*20 + ['SS']*20 + 
           ['LF']*20 + ['CF']*20 + ['RF']*20 + ['DH']*20 + ['P']*20,
    'SB': np.concatenate([
        np.random.poisson(3, 20),    # C - few stolen bases
        np.random.poisson(4, 20),    # 1B - few stolen bases
        np.random.poisson(15, 20),   # 2B - moderate stolen bases
        np.random.poisson(10, 20),   # 3B - moderate stolen bases
        np.random.poisson(20, 20),   # SS - high stolen bases
        np.random.poisson(12, 20),   # LF - moderate stolen bases
        np.random.poisson(25, 20),   # CF - highest stolen bases
        np.random.poisson(12, 20),   # RF - moderate stolen bases
        np.random.poisson(5, 20),    # DH - few stolen bases
        np.random.poisson(1, 20)     # P - very few stolen bases
    ]),
    'CS': np.random.binomial(n=10, p=0.3, size=200),  # Caught stealing attempts
    'Speed': np.random.normal(27, 2, 200),  # Speed score (ft/sec)
    'PA': np.random.normal(500, 50, 200)    # Plate appearances
}

# Calculate additional metrics
df = pd.DataFrame(data)
df['SB_Success_Rate'] = (df['SB'] / (df['SB'] + df['CS'])) * 100
df['SB_per_PA'] = (df['SB'] / df['PA']) * 100
```

```python
# Create multiple visualizations
plt.figure(figsize=(15, 10))

# 1. Stolen Bases by Position (Box Plot)
plt.subplot(2, 2, 1)
sns.boxplot(data=df[df['Pos'] != 'P'], x='Pos', y='SB', 
            order=['SS', 'CF', '2B', 'LF', 'RF', '3B', '1B', 'C', 'DH'])
plt.title('Stolen Bases Distribution by Position')
plt.xticks(rotation=45)
plt.ylabel('Stolen Bases')

# 2. Success Rate by Position
plt.subplot(2, 2, 2)
sns.boxplot(data=df[df['Pos'] != 'P'], x='Pos', y='SB_Success_Rate',
            order=['SS', 'CF', '2B', 'LF', 'RF', '3B', '1B', 'C', 'DH'])
plt.title('Stolen Base Success Rate by Position')
plt.xticks(rotation=45)
plt.ylabel('Success Rate (%)')

# 3. Speed vs Stolen Bases
plt.subplot(2, 2, 3)
sns.scatterplot(data=df[df['Pos'] != 'P'], x='Speed', y='SB', hue='Pos', alpha=0.6)
plt.title('Speed vs Stolen Bases')
plt.xlabel('Speed Score (ft/sec)')
plt.ylabel('Stolen Bases')

# 4. Stolen Bases per Plate Appearance
plt.subplot(2, 2, 4)
sns.boxplot(data=df[df['Pos'] != 'P'], x='Pos', y='SB_per_PA',
            order=['SS', 'CF', '2B', 'LF', 'RF', '3B', '1B', 'C', 'DH'])
plt.title('Stolen Bases per 100 Plate Appearances')
plt.xticks(rotation=45)
plt.ylabel('SB per 100 PA')

plt.tight_layout()
plt.show()
```

```python
# Print summary statistics
print("Stolen Base Statistics by Position:")
summary_stats = df[df['Pos'] != 'P'].groupby('Pos').agg({
    'SB': ['mean', 'std', 'max'],
    'SB_Success_Rate': 'mean',
    'Speed': 'mean'
}).round(2)

print(summary_stats)
```

### Key Observations:

1. **Position Trends:**
   - Center Fielders and Shortstops lead in stolen base attempts
   - Catchers and First Basemen show lowest stolen base numbers
   - Middle infielders (2B, SS) show consistent stealing activity

2. **Success Rates:**
   - Faster positions tend to have higher success rates
   - Speed shows strong correlation with stolen base totals
   - Success rates vary more for positions that steal less frequently

3. **Efficiency:**
   - Some positions show high totals but lower efficiency
   - Speed isn't the only factor in stolen base success
   - Playing time (PA) affects overall stolen base opportunities
